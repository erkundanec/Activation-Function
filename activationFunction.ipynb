# Binary step function

```python
import matplotlib.pyplot as plt
import numpy as np

def step(x):
    length = len(x)
    t = x.copy()
    ds = x.copy()
    for i in np.arange(length):
        if x[i]<0:
            t[i] = 0
            ds[i] = 0
        else:
            t[i] = 1
            ds[i] = 0 
    return t,ds

x=np.arange(-8,8,0.01)
relu(x)
# Setup centered axes
fig, ax = plt.subplots(figsize=(7, 4))
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

ax.spines['left'].set_position('center')
# ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# Create and show plot
ax.plot(x,step(x)[0], linewidth=2, label="binary Step")
ax.plot(x,step(x)[1],  linewidth=2, label="derivative")
plt.xlim([-8, +8])
plt.ylim([-0.5, 1.1])
# plt.text(-8, 0.8, r'$f(z)=\frac{e^{z}-e^{-z}}{e^{z}+e^{-z}}$',fontsize=20)
plt.text(2.5, 0.45, r'$f(z) = 1~~~if~ z\geq 0$',fontsize=20)
plt.text(2.5, 0.32, r'$f(z)=0,~{otherwise}$',fontsize=20)
plt.text(2.5, 0.12, r'$f^\prime(z) = 0,~~{for}~all~z$',fontsize=20)

plt.ylabel(r'$f(z)$',horizontalalignment='right',y=1.0)
plt.xlabel(r'$z$',horizontalalignment='right',x=1)
ax.legend(loc="upper right", frameon=False)
plt.axis('tight')
plt.savefig('BinaryStep.pdf')
```

# Sigmoid Function

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
    s=1/(1+np.exp(-x))
    ds=s*(1-s)  
    return s,ds

x=np.arange(-8,8,0.01)
sigmoid(x)
# Setup centered axes
fig, ax = plt.subplots(figsize=(7, 4))
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# Create and show plot
ax.plot(x,sigmoid(x)[0], linewidth=2, label="sigmoid")
ax.plot(x,sigmoid(x)[1],  linewidth=2, label="derivative")
plt.xlim([-8, +8])
plt.ylim([-0.5, 1.1])
plt.text(2.5, 0.4, r'$f(z)=\frac{1}{1+e^{-z}}$',fontsize=20)
plt.text(2.5, 0.25, r'$f^\prime(z)=f(z)(1-f(z))$',fontsize=20)
plt.ylabel(r'$f(z)$',horizontalalignment='right',y=1.0)
plt.xlabel(r'$z$',horizontalalignment='right',x=1)
ax.legend(loc="upper right", frameon=False)
plt.axis('tight')
plt.savefig('sigmoid.pdf')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def tanh(x):
    t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    dt=1-t**2
    return t,dt

x=np.arange(-8,8,0.01)
tanh(x)
# Setup centered axes
fig, ax = plt.subplots(figsize=(7, 4))
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# Create and show plot
ax.plot(x,tanh(x)[0], linewidth=2, label="tanh")
ax.plot(x,tanh(x)[1],  linewidth=2, label="derivative")
plt.xlim([-8, +8])
plt.ylim([-0.5, 1.1])
plt.text(-8, 0.5, r'$f(z)=\frac{e^{z}-e^{-z}}{e^{z}+e^{-z}}$',fontsize=20)
plt.text(-8, 0.2, r'$f^\prime(z)=1-f^2(z)$',fontsize=20)
plt.ylabel(r'$f(z)$',horizontalalignment='right',y=1.0)
plt.xlabel(r'$z$',horizontalalignment='right',x=1)
ax.legend(loc="upper right", frameon=False)
plt.axis('tight')
plt.savefig('tanh.pdf')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def relu(x):
    t = np.maximum(0,x)
    return t

x=np.arange(-8,8,0.01)
relu(x)
# Setup centered axes
fig, ax = plt.subplots(figsize=(7, 4))
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

ax.spines['left'].set_position('center')
# ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# Create and show plot
ax.plot(x,relu(x), linewidth=2, label="ReLU")
# ax.plot(x,tanh(x)[1],  linewidth=2, label="derivative")
plt.xlim([-8, +8])
plt.ylim([-0.5, 1.1])
# plt.text(-8, 0.8, r'$f(z)=\frac{e^{z}-e^{-z}}{e^{z}+e^{-z}}$',fontsize=20)
plt.text(-8, 4.3, r'$f(z)=\max(0,z)$',fontsize=20)
plt.text(-8.3, 3, r'$f^{\prime}(x) = 1, ~~x>=0$',fontsize=20)
plt.text(-6.3, 2.2, r'$= 0,~~ x<0$',fontsize=20)

      

plt.ylabel(r'$f(z)$',horizontalalignment='right',y=1.0)
plt.xlabel(r'$z$',horizontalalignment='right',x=1)
ax.legend(loc="upper right", frameon=False)
plt.axis('tight')
plt.savefig('relu.pdf')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def relu(x):
    length = len(x)
    t = x.copy()
    for i in np.arange(length):
        if x[i]<0:
            t[i] = 0.1*x[i]
        else:
            t[i] = x[i]
    return t

x=np.arange(-8,8,0.01)
relu(x)
# Setup centered axes
fig, ax = plt.subplots(figsize=(9, 5))
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# Create and show plot
ax.plot(x,relu(x), linewidth=2, label="Leaky ReLU")
# ax.plot(x,tanh(x)[1],  linewidth=2, label="derivative")
plt.xlim([-8, +8])
plt.ylim([-0.5, 1.1])
# plt.text(-8, 0.8, r'$f(z)=\frac{e^{z}-e^{-z}}{e^{z}+e^{-z}}$',fontsize=20)
plt.text(-8, 2, r'$f(z)=az,~z<0$',fontsize=20)
plt.text(-8, 1, r'$f(z)=z,~{otherwise}$',fontsize=20)
plt.ylabel(r'$f(z)$',horizontalalignment='right',y=1.0)
plt.xlabel(r'$z$',horizontalalignment='right',x=1)
ax.legend(loc="upper right", frameon=False)
plt.axis('tight')
plt.savefig('leakyRelu.pdf')
```
